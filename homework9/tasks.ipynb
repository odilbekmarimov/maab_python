```python
import sqlite3
import pandas as pd
```

```python
with sqlite3.connect('population.db') as con:
    df = pd.read_sql("select * from population", con = con)
```

# 1

```python
task1 = df.loc[(df['gender'] == 'Female') & (df['state'] == 'Florida'), 
    ['first_name', 'last_name', 'salary', 'gender', 'state']]
task1 = task1.rename(columns={
    'first_name': 'FirstName',
    'last_name': 'LastName',
    'salary': 'Salary',
    'gender': 'Gender',
    'state': 'State'
})
task1
```
